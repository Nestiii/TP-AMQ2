Test de Integración de la API

Este notebook verifica que el servicio de predicción esté correctamente levantado y consumiendo el modelo Champion desde MLflow.

```python
import requests
import pandas as pd

BASE_URL = "http://localhost:8800"

print(f"Probando conexión con la API en: {BASE_URL}")
```

```python
def test_health():
    response = requests.get(f"{BASE_URL}/")
    if response.status_code == 200:
        data = response.json()
        print("API Online")
        print(f"Modelo: {data.get('model')}")
        print(f"Versión: {data.get('model_version')}")
    else:
        print(f"Error de conexión: {response.status_code}")

test_health()
```

```python
def test_prediction():

    payload = {
        "features": {
            "f": 1000,
            "alpha": 5.4,
            "c": 0.1524,
            "U_infinity": 39.6,
            "delta": 0.00529
        }
    }
    
    response = requests.post(f"{BASE_URL}/predict/", json=payload)
    
    if response.status_code == 200:
        result = response.json()
        print("Predicción exitosa")
        print(f"Nivel de presión sonora (SSPL): {result.get('sspl_db'):.2f} dB")
        print(f"Usando modelo versión: {result.get('model_version')}")
    else:
        print(f"Fallo en la predicción: {response.text}")

test_prediction()
```
